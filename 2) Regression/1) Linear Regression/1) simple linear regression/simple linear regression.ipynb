```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading Data
data = pd.read_csv('data/Salary_Data.csv')
data.head()
```

```python
# Select DV & ID
X = data.iloc[:, :-1]
Y = data.iloc[:,-1]
```

```python
# general exploration of the data
data.info()
```

```python
# split data into train / test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=.2,random_state=0)
```

```python
# build Linear regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

# Visualization

```python
# visualize the accuracy of training model
plt.scatter(x_train,y_train,color='red') # plot actual data
plt.plot(x_train,regressor.predict(x_train)) # plot the model line
plt.title('Years of Experience VS Salaries (training model)')
plt.xlabel('Years of Experience')
plt.ylabel('Salaries')
plt.show()
```

```python
# visualize the accuracy of testing model
plt.scatter(x_test,y_test,color='red') # plot actual data
plt.plot(x_train,regressor.predict(x_train)) # plot  the model line (the same like above without change)
plt.title('Years of Experience VS Salaries (testing model)')
plt.xlabel('Years of Experience')
plt.ylabel('Salaries')
plt.show()
```

# Improve of the visualization Code Style

```python
def visualize (actual_X,actual_Y):
    '''This function for visualizing the accuracy of the model for the training or testing'''
    plt.scatter(actual_X,actual_Y,color='red') # ploting the actual values
    plt.plot(x_train,regressor.predict(x_train),color='blue') # plotting the model (we didn't use the paramters)
    plt.title('Years of Experience VS Salaries (testing model)')
    plt.xlabel('Years of Experience')
    plt.ylabel('Salaries')
    plt.show()
```

```python
# visualize the accuracy of training model
visualize(x_train,y_train)
```

```python
# visualize the accuracy of testing model
visualize(x_test,y_test)
```
